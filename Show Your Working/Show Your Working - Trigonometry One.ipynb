# Show Your Working - Trigonometry One

<hr>

Here are some trigonometry exercise to help you with your studies. The answers follow after the questions.

## Question 1
<hr>
Prove that:
$$\cos^2\theta-\sin^2\theta=2\cos^2\theta-1$$
## Question 2
<hr>
Prove that:
$$\sin^2\theta+2\cos^2\theta=2-\sin^2\theta$$
## Question 3
<hr>
Prove that:
$$(\sin\theta+\cos\theta)^2-2\sin\theta\cos\theta=1$$
## Question 4
<hr>
Prove that:
$$\sin^3\theta+\cos^3\theta=(\sin\theta+\cos\theta)(1-\sin\theta\cos\theta)$$
## Question 5
<hr>
Prove that:
$$\cos^2\theta+3\sin^2\theta=3-2\cos^2\theta$$
## Question 6
<hr>
Prove that:
$$\cos^4\theta-\sin^4\theta=\cos^2\theta-\sin^2\theta$$
## Question 7
<hr>
Prove that:
$$\frac{\cos^2\theta}{1-\sin\theta}=1+\sin\theta$$
## Question 8
<hr>
Prove that:
$$\frac{\sin\theta}{1+\cos\theta}=\frac{1-\cos\theta}{\sin\theta}$$
## Question 9
<hr>
Prove that:
$$\frac{\sin\theta}{1+\cos\theta}+\frac{1+\cos\theta}{\sin\theta}=\frac{2}{\sin\theta}$$
## Question 10
<hr>
Prove that:
$$(\cos\theta+\sin\theta)^2+(\cos\theta-\sin\theta)^2=2$$

<br>
<hr><hr>
<br>

## Answer to Question 1
<hr>
Using the identity:
$$\sin^2\theta=1-\cos^2\theta$$<br>
Proof:
$$\cos^2\theta-\sin^2\theta$$<br>
$$=\cos^2\theta-(1-\cos^2\theta)$$<br>
$$=\cos^2\theta-1+\cos^2\theta$$<br>
$$=2\cos^2\theta-1$$<br>
## Answer to Question 2
<hr>
Using the identity:
$$\cos^2\theta=1-\sin^2\theta$$<br>
Proof:
$$\sin^2\theta+2\cos^2\theta$$<br>
$$=\sin^2\theta+2(1-\sin^2\theta)$$<br>
$$=\sin^2\theta+2-2\sin^2\theta$$<br>
$$=2-\sin^2\theta$$<br>
## Answer to Question 3
<hr>
Proof:
$$(\sin\theta+\cos\theta)^2-2\sin\theta\cos\theta$$<br>
$$=\sin^2\theta+2\cos\theta\sin\theta+\cos^2\theta-2\sin\theta\cos\theta$$<br>
$$=\sin^2\theta+\cos^2\theta$$<br>
$$=1$$<br>
## Answer to Question 4
<hr>
Proof:
$$(\sin\theta+\cos\theta)(1-\sin\theta\cos\theta)$$<br>
$$=\sin\theta+\cos\theta-(\sin^2\theta\cos\theta+\sin\theta\cos^2\theta)$$<br>
$$=\sin\theta+\cos\theta-((1-\cos^2\theta)\cos\theta+\sin\theta(1-\sin^2\theta))$$<br>
$$=\sin\theta+\cos\theta - \cos\theta + \cos^3\theta - \sin\theta + \sin^3\theta$$<br>
$$=\cos^3\theta+\sin^3\theta$$<br>
## Answer to Question 5
<hr>
Proof:
$$\cos^2\theta+3\sin^2\theta$$<br>
$$=\cos^2\theta + 3(1-\cos^2\theta)$$<br>
$$=\cos^2\theta+3-3\cos^2\theta$$<br>
$$=3-2\cos^2\theta$$<br>
## Answer to Question 6
<hr>
Proof:
$$\cos^4\theta-\sin^4\theta$$<br>
$$=(\cos^2\theta+\sin^2\theta)(\cos^2\theta-\sin^2\theta)$$<br>
$$=(1)(\cos^2\theta-\sin^2\theta)$$<br>
$$=\cos^2\theta-\sin^2\theta$$<br>
## Answer to Question 7
<hr>
Proof:
$$\frac{\cos^2\theta}{1-\sin\theta}$$<br>
$$=\frac{1-\sin^2\theta}{1-\sin\theta}$$<br>
$$=\frac{(1-\sin\theta)(1+\sin\theta)}{1-\sin\theta}$$<br>
$$=1+\sin\theta$$<br>
## Answer to Question 8
<hr>
Proof:
$$\frac{\sin\theta}{1+\cos\theta}$$<br>
$$=\frac{\sin\theta}{1+\cos\theta}\times\frac{1-\cos\theta}{1-\cos\theta}$$<br>
$$=\frac{\sin\theta(1-\cos\theta)}{(1+\cos\theta-\cos\theta+\cos^2\theta)}$$<br>
$$=\frac{\sin\theta(1-\cos\theta)}{(1+\cos^2\theta)}$$<br>
$$=\frac{\sin\theta(1-\cos\theta)}{(\sin^2\theta)}$$<br>
$$=\frac{1-\cos\theta}{\sin\theta}$$<br>
## Answer to Question 9
<hr>
Proof:
$$\frac{\sin\theta}{1+\cos\theta}+\frac{1+\cos\theta}{\sin\theta}$$<br>
$$=\frac{\sin\theta}{1+\cos\theta}\frac{\sin\theta}{\sin\theta}+\frac{1+\cos\theta}{\sin\theta}\frac{1+\cos\theta}{1+\cos\theta}$$<br>
$$=\frac{\sin^2\theta+1+\cos^2\theta+2\cos\theta}{(1+\cos\theta)\sin\theta}$$<br>
$$=\frac{2+2\cos\theta}{(1+\cos\theta)\sin\theta)}$$<br>
$$=\frac{2(1+\cos\theta)}{(1+\cos\theta)\sin\theta}$$<br>
$$=\frac{2}{\sin\theta}$$<br>
## Answer to Question 10
<hr>
Proof:
$$(\cos\theta+\sin\theta)^2+(\cos\theta-\sin\theta)^2$$<br>
$$=\cos^2\theta+\sin^2\theta+2\cos\theta\sin\theta+\cos^2\theta+\sin^2\theta-2\sin\theta\cos\theta$$<br>
$$=2(\cos^2\theta+\sin^2\theta)$$<br>
$$=2$$<br>
