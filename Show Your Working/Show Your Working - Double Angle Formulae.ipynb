## Question 1

<hr>

Prove that:

$$ \mathrm{cosec}2\alpha = \mathrm{cosec}\alpha\sec\alpha $$

## Question 2

<hr>

Prove that:

$$ \tan\alpha + \cot\alpha = 2\mathrm{cosec}2\alpha $$

## Question 3

<hr>

Prove that:

$$ \frac{1+\tan^2\alpha}{1-\tan^2\alpha} = \sec2\alpha $$

## Question 4

<hr>

Prove that:

$$ \cot2\alpha = \mathrm{cosec}2\alpha - \tan\alpha $$

## Question 5

<hr>

Prove that:

$$ \frac{\sin2\alpha}{1-\cos^2\alpha} = 2\cot\alpha $$

## Question 6

<hr>

Prove that:

$$ \tan\alpha - \cot\alpha = -2\cot\alpha $$

## Question 7

<hr>

Prove that:

$$ \mathrm{cosec}2\alpha + \cot2\alpha = \cot\alpha $$

## Question 8

<hr>

Prove that:

$$ \tan(\frac{\alpha}{2}) = \frac{\sin\alpha}{1+\cos\alpha} $$

## Question 9

<hr>

Prove that:

$$ \sin\alpha = \frac{2\tan(\frac{\alpha}{2})}{1+\tan^2(\frac{\alpha}{2})} $$

<br><hr><hr><br>

## Answer to Question 1
<hr>
Proof:

$$ 2\mathrm{cosec}2\alpha $$<br>
$$ =\frac{2}{\sin2\alpha} $$<br>
$$ =\frac{2}{2\sin\alpha\cos\alpha} $$<br>
$$ =\frac{1}{\sin\alpha\cos\alpha} $$<br>
$$ =\mathrm{cosec}\alpha\sec\alpha $$


## Answer to Question 2
<hr>
Proof:

$$ 2\mathrm{cosec}2\alpha $$

$$ =\frac{2}{\sin2\alpha} $$

$$ =\frac{2}{2\sin\alpha\cos\alpha} $$

$$ =\frac{1}{\sin\alpha\cos\alpha} $$

$$ =\frac{\sin^2\alpha+\cos^2\alpha}{\sin\alpha\cos\alpha} $$

$$ =\frac{\sin^2\alpha}{\sin\alpha\cos\alpha} + \frac{\cos^2\alpha}{\sin\alpha\cos\alpha} $$

$$ =\frac{\sin\alpha}{\cos\alpha} + \frac{\cos\alpha}{\sin\alpha} $$

$$ =\tan\alpha +\cot\alpha $$ 


## Answer to Question 3
<hr>
Proof:

$$ \sec^2\alpha $$

$$ =\frac{1}{cos2\alpha} $$

$$ =\frac{1}{cos^2\alpha-\sin^2\alpha} $$

$$ =\frac{cos^2\alpha+\sin^2\alpha}{cos^2\alpha-\sin^2\alpha} $$

Now, we know $\tan\alpha=\frac{\sin\alpha}{\cos\alpha}$ so lets divide everything by $\cos\alpha$

$$ =\frac{\frac{c\os^2\alpha}{\cos^2\alpha}+\frac{\sin^2\alpha}{\cos^2\alpha}}{\frac{\cos^2\alpha}{\cos^2\alpha}-\frac{\sin^2\alpha}{\cos^2\alpha}} $$

$$ = \frac{1+\tan^2\alpha}{1-\tan^2\alpha} $$


## Answer to Question 4
<hr>
Proof:

$$ \cot2\alpha = \mathrm{cosec}2\alpha - \tan\alpha $$

$$ \frac{\cot^2\alpha-1}{2\cot\alpha} = \frac{1}{2\sin\alpha\cos\alpha} - \frac{\sin\alpha}{\cos\alpha} $$

Taking the right hand side and knowing we want to get $2\cot\alpha$ on the deonminator, we shall divide $\frac{1}{2\sin\alpha\cos\alpha}$ by $\sin^2\alpha$ and $\frac{\sin\alpha}{\cos\alpha}$ by $\frac{2}{\sin\alpha}$


$$ = \frac{\frac{1}{\sin^2\alpha}}{\frac{2\sin\alpha\cos\alpha}{\sin^2\alpha}} - \frac{2\frac{\sin\alpha}{\sin\alpha}}{2\frac{cos\alpha}{\sin\alpha}} $$

$$ = \frac{\frac{1}{\sin^2\alpha}}{2\cot\alpha} - \frac{2}{2\cot\alpha} $$

$$ = \frac{\mathrm{cosec^2\alpha-2}}{2\cot\alpha} $$

$$ = \frac{\cot^2\alpha+1-2}{2\cot\alpha} $$

$$ = \frac{\cot^2\alpha-1}{2\cot\alpha} $$


## Answer to Question 5
<hr>
Proof:

$$ \frac{\sin2\alpha}{1-\cos^2\alpha} $$

$$ = \frac{2\sin\alpha\cos\alpha}{1-\cos^2\alpha} $$

$$ = \frac{2\sin\alpha\cos\alpha}{\sin^2\alpha}  $$

$$ = \frac{2\cos\alpha}{\sin\alpha}  $$

$$ = 2\cot\alpha  $$


## Answer to Question 6
<hr>
Proof:

$$ -2\cot\alpha $$

$$ = \frac{-2(\cot^2\alpha-1)}{2\cot\alpha} $$

$$ = \frac{1-\cot^2\alpha}{\cot\alpha} $$

$$ = \frac{1}{\cot\alpha} - \cot\alpha $$

$$ = \tan\alpha - \cot\alpha $$


## Answer to Question 7
<hr>
Proof:

$$ \mathrm{cosec}2\alpha + \cot2\alpha $$

$$ = \frac{1}{\sin\alpha\cos\alpha} + \frac{\cot^2\alpha-1}{2\cot\alpha} $$

$$ = \frac{\frac{1}{\sin^2\alpha}}{\frac{\sin\alpha\cos\alpha}{\sin^2\alpha}} + \frac{\cot^2\alpha-1}{2\cot\alpha} $$

$$ = \frac{\mathrm{cosec}^2\alpha}{2\cot\alpha} + \frac{\cot^2\alpha-1}{2\cot\alpha} $$

$$ = \frac{\mathrm{cosec}^2\alpha+\cot^2\alpha-1}{2\cot\alpha} $$

$$ = \frac{\cot^2\alpha+1+\cot^2\alpha-1}{2\cot\alpha} $$

$$ = \frac{2\cot^2\alpha}{2\cot\alpha} $$

$$ = 2\cot\alpha $$


## Answer to Question 8
<hr>
Proof:

$$ \frac{\sin\alpha}{1+\cos\alpha} $$

$$ = \frac{\sin(2\frac{\alpha}{2})}{1+\cos(2\frac{\alpha}{2})} $$

$$ = \frac{2\sin(\frac{\alpha}{2})\cos(\frac{\alpha}{2})}{1-2\cos^2(\frac{\alpha}{2})-1} $$

$$ = \frac{2\sin(\frac{\alpha}{2})\cos(\frac{\alpha}{2})}{2\cos^2(\frac{\alpha}{2})} $$

$$ = \frac{\sin(\frac{\alpha}{2})}{\cos(\frac{\alpha}{2})} $$

$$ = \tan(\frac{\alpha}{2}) $$


## Answer to Question 9
<hr>
Proof:

$$ \sin\alpha $$

$$ = \sin(2\frac{\alpha}{2}) $$

$$ = 2\sin(\frac{\alpha}{2})\cos(\frac{\alpha}{2}) $$

$$ = \frac{\frac{2\sin(\frac{\alpha}{2})\cos(\frac{\alpha}{2})}{\cos^2(\frac{\alpha}{2})}}{\frac{1}{\cos^2(\frac{\alpha}{2})}} $$

$$ = \frac{2\tan(\frac{\alpha}{2})}{\sec^2(\frac{\alpha}{2})} $$

$$ = \frac{2\tan(\frac{\alpha}{2})}{1+\tan^2(\frac{\alpha}{2})} $$
