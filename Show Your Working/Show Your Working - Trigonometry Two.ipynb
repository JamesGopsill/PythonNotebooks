# Show Your Working - Trigonometry Two

<hr>


## Question 11
<hr>
Prove that:
$$\cos^4\theta-\sin^4\theta+1=2\cos^2\theta$$
## Question 12
<hr>
Prove that:
$$\tan\theta\mathrm{cosec}\theta=\sec\theta$$
## Question 13
<hr>
Prove that:
$$\frac{1}{\sec^2\theta}+\frac{1}{\mathrm{cosec}^2\theta}=1$$
## Question 14
<hr>
Prove that:
$$\sec\theta-\cos\theta=\sin\theta\tan\theta$$
## Question 15
<hr>
Prove that:
$$\mathrm{cosec}\theta-\sin\theta=\cot\theta\cos\theta$$
## Question 16
<hr>
Prove that:
$$\cos^2\theta-\sin^2\theta=2\cos^2\theta-1$$
## Question 17
<hr>
Prove that:
$$\sin^2\theta+2\cos^2\theta=2-\sin^2\theta$$
## Question 18
<hr>
Prove that:
$$(\sin\theta+\cos\theta)^2-2\sin\theta\cos\theta=1$$
## Question 19
<hr>
Prove that:
$$\sin^3\theta+\cos^3\theta=(\sin\theta+\cos\theta)(1-\sin\theta\cos\theta)$$
## Question 20
<hr>
Prove that:
$$\cos^2\theta+3\sin^2\theta=3-2\cos^2\theta$$

<br><hr><hr><br>

## Answer to Question 11
<hr>
Proof:
$$\cos^4\theta-\sin^4\theta+1$$<br>
$$=\cos^4\theta-(\sin^2\theta)(\sin^2\theta)+1$$<br>
$$=\cos^4\theta-(1-\cos^2\theta)(1-\cos^2\theta)+1$$<br>
$$=\cos^4\theta-(\cos^4\theta-2\cos\theta+1)+1$$<br>
$$=2\cos^2\theta$$<br>
## Answer to Question 12
<hr>
Proof:
$$\tan\theta\mathrm{cosec}\theta$$<br>
$$=\frac{\sin\theta}{\cos\theta}\frac{1}{\sin\theta}$$<br>
$$=\frac{1}{\cos\theta}$$<br>
$$=\sec\theta$$<br>
## Answer to Question 13
<hr>
Proof:
$$\frac{1}{\sec^2\theta}+\frac{1}{\mathrm{cosec}^2\theta}$$<br>
$$=\frac{1}{\frac{1}{\cos^2\theta}}+\frac{1}{\frac{1}{\sin^2\theta}}$$<br>
$$=\cos^2\theta+\sin^2\theta$$<br>
$$=1$$<br>
## Answer to Question 14
<hr>
Proof:
$$\sec\theta-\cos\theta=\sin\theta\tan\theta$$<br>
$$\frac{1}{\cos\theta}-\cos\theta=\sin\theta\frac{\sin\theta}{\cos\theta}$$<br>
Taking the right-hand side of the equation.
$$=\frac{\sin^2\theta}{\cos\theta}$$<br>
$$=\frac{1-\cos^2\theta}{\cos\theta}$$<br>
$$=\frac{1}{\cos\theta}-\frac{\cos^2\theta}{\cos\theta}$$<br>
$$=\frac{1}{\cos\theta}-\cos\theta$$<br>
## Answer to Question 15
<hr>
Proof:
$$\mathrm{cosec}\theta-\sin\theta=\cot\theta\cos\theta$$<br>
$$\frac{1}{\sin\theta}-\sin\theta=\frac{\cos\theta}{\sin\theta}\cos\theta$$<br>
Taking the right-hand side of the equation.
$$=\frac{\cos^2\theta}{\sin\theta}$$<br>
$$=\frac{1-\sin^2\theta}{\sin\theta}$$<br>
$$=\frac{1}{\sin\theta}-\frac{\sin^2\theta}{\sin\theta}$$<br>
$$=\frac{1}{\sin\theta}-\sin\theta$$<br>
## Answer to Question 16
<hr>
Using the identity:
$$\sin^2\theta=1-\cos^2\theta$$<br>
Proof:
$$\cos^2\theta-\sin^2\theta$$<br>
$$=\cos^2\theta-(1-\cos^2\theta)$$<br>
$$=\cos^2\theta-1+\cos^2\theta$$<br>
$$=2\cos^2\theta-1$$<br>
## Answer to Question 17
<hr>
Using the identity:
$$\cos^2\theta=1-\sin^2\theta$$<br>
Proof:
$$\sin^2\theta+2\cos^2\theta$$<br>
$$=\sin^2\theta+2(1-\sin^2\theta)$$<br>
$$=\sin^2\theta+2-2\sin^2\theta$$<br>
$$=2-\sin^2\theta$$<br>
## Answer to Question 18
<hr>
Proof:
$$(\sin\theta+\cos\theta)^2-2\sin\theta\cos\theta$$<br>
$$=\sin^2\theta+2\cos\theta\sin\theta+\cos^2\theta-2\sin\theta\cos\theta$$<br>
$$=\sin^2\theta+\cos^2\theta$$<br>
$$=1$$<br>
## Answer to Question 19
<hr>
Proof:
$$(\sin\theta+\cos\theta)(1-\sin\theta\cos\theta)$$<br>
$$=\sin\theta+\cos\theta-(\sin^2\theta\cos\theta+\sin\theta\cos^2\theta)$$<br>
$$=\sin\theta+\cos\theta-((1-\cos^2\theta)\cos\theta+\sin\theta(1-\sin^2\theta))$$<br>
$$=\sin\theta+\cos\theta - \cos\theta + \cos^3\theta - \sin\theta + \sin^3\theta$$<br>
$$=\cos^3\theta+\sin^3\theta$$<br>
## Answer to Question 20
<hr>
Proof:
$$\cos^2\theta+3\sin^2\theta$$<br>
$$=\cos^2\theta + 3(1-\cos^2\theta)$$<br>
$$=\cos^2\theta+3-3\cos^2\theta$$<br>
$$=3-2\cos^2\theta$$<br>
