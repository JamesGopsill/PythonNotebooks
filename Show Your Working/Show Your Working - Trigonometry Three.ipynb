# Show Your Working - Trigonometry Three

<hr>


## Question 21
<hr>
Prove that:
$$\cos^4\theta-\sin^4\theta=\cos^2\theta-\sin^2\theta$$
## Question 22
<hr>
Prove that:
$$\frac{\cos^2\theta}{1-\sin\theta}=1+\sin\theta$$
## Question 23
<hr>
Prove that:
$$\frac{\sin\theta}{1+\cos\theta}=\frac{1-\cos\theta}{\sin\theta}$$
## Question 24
<hr>
Prove that:
$$\frac{\sin\theta}{1+\cos\theta}+\frac{1+\cos\theta}{\sin\theta}=\frac{2}{\sin\theta}$$
## Question 25
<hr>
Prove that:
$$(\cos\theta+\sin\theta)^2+(\cos\theta-\sin\theta)^2=2$$
## Question 26
<hr>
Prove that:
$$\cos^4\theta-\sin^4\theta+1=2\cos^2\theta$$
## Question 27
<hr>
Prove that:
$$\tan\theta\mathrm{cosec}\theta=\sec\theta$$
## Question 28
<hr>
Prove that:
$$\frac{1}{\sec^2\theta}+\frac{1}{\mathrm{cosec}^2\theta}=1$$
## Question 29
<hr>
Prove that:
$$\sec\theta-\cos\theta=\sin\theta\tan\theta$$
## Question 30
<hr>
Prove that:
$$\mathrm{cosec}\theta-\sin\theta=\cot\theta\cos\theta$$

<br><hr><hr><br>

## Answer to Question 21
<hr>
Proof:
$$\cos^4\theta-\sin^4\theta$$<br>
$$=(\cos^2\theta+\sin^2\theta)(\cos^2\theta-\sin^2\theta)$$<br>
$$=(1)(\cos^2\theta-\sin^2\theta)$$<br>
$$=\cos^2\theta-\sin^2\theta$$<br>
## Answer to Question 22
<hr>
Proof:
$$\frac{\cos^2\theta}{1-\sin\theta}$$<br>
$$=\frac{1-\sin^2\theta}{1-\sin\theta}$$<br>
$$=\frac{(1-\sin\theta)(1+\sin\theta)}{1-\sin\theta}$$<br>
$$=1+\sin\theta$$<br>
## Answer to Question 23
<hr>
Proof:
$$\frac{\sin\theta}{1+\cos\theta}$$<br>
$$=\frac{\sin\theta}{1+\cos\theta}\times\frac{1-\cos\theta}{1-\cos\theta}$$<br>
$$=\frac{\sin\theta(1-\cos\theta)}{(1+\cos\theta-\cos\theta+\cos^2\theta)}$$<br>
$$=\frac{\sin\theta(1-\cos\theta)}{(1+\cos^2\theta)}$$<br>
$$=\frac{\sin\theta(1-\cos\theta)}{(\sin^2\theta)}$$<br>
$$=\frac{1-\cos\theta}{\sin\theta}$$<br>
## Answer to Question 24
<hr>
Proof:
$$\frac{\sin\theta}{1+\cos\theta}+\frac{1+\cos\theta}{\sin\theta}$$<br>
$$=\frac{\sin\theta}{1+\cos\theta}\frac{\sin\theta}{\sin\theta}+\frac{1+\cos\theta}{\sin\theta}\frac{1+\cos\theta}{1+\cos\theta}$$<br>
$$=\frac{\sin^2\theta+1+\cos^2\theta+2\cos\theta}{(1+\cos\theta)\sin\theta}$$<br>
$$=\frac{2+2\cos\theta}{(1+\cos\theta)\sin\theta)}$$<br>
$$=\frac{2(1+\cos\theta)}{(1+\cos\theta)\sin\theta}$$<br>
$$=\frac{2}{\sin\theta}$$<br>
## Answer to Question 25
<hr>
Proof:
$$(\cos\theta+\sin\theta)^2+(\cos\theta-\sin\theta)^2$$<br>
$$=\cos^2\theta+\sin^2\theta+2\cos\theta\sin\theta+\cos^2\theta+\sin^2\theta-2\sin\theta\cos\theta$$<br>
$$=2(\cos^2\theta+\sin^2\theta)$$<br>
$$=2$$<br>
## Answer to Question 26
<hr>
Proof:
$$\cos^4\theta-\sin^4\theta+1$$<br>
$$=\cos^4\theta-(\sin^2\theta)(\sin^2\theta)+1$$<br>
$$=\cos^4\theta-(1-\cos^2\theta)(1-\cos^2\theta)+1$$<br>
$$=\cos^4\theta-(\cos^4\theta-2\cos\theta+1)+1$$<br>
$$=2\cos^2\theta$$<br>
## Answer to Question 27
<hr>
Proof:
$$\tan\theta\mathrm{cosec}\theta$$<br>
$$=\frac{\sin\theta}{\cos\theta}\frac{1}{\sin\theta}$$<br>
$$=\frac{1}{\cos\theta}$$<br>
$$=\sec\theta$$<br>
## Answer to Question 28
<hr>
Proof:
$$\frac{1}{\sec^2\theta}+\frac{1}{\mathrm{cosec}^2\theta}$$<br>
$$=\frac{1}{\frac{1}{\cos^2\theta}}+\frac{1}{\frac{1}{\sin^2\theta}}$$<br>
$$=\cos^2\theta+\sin^2\theta$$<br>
$$=1$$<br>
## Answer to Question 29
<hr>
Proof:
$$\sec\theta-\cos\theta=\sin\theta\tan\theta$$<br>
$$\frac{1}{\cos\theta}-\cos\theta=\sin\theta\frac{\sin\theta}{\cos\theta}$$<br>
Taking the right-hand side of the equation.
$$=\frac{\sin^2\theta}{\cos\theta}$$<br>
$$=\frac{1-\cos^2\theta}{\cos\theta}$$<br>
$$=\frac{1}{\cos\theta}-\frac{\cos^2\theta}{\cos\theta}$$<br>
$$=\frac{1}{\cos\theta}-\cos\theta$$<br>
## Answer to Question 30
<hr>
Proof:
$$\mathrm{cosec}\theta-\sin\theta=\cot\theta\cos\theta$$<br>
$$\frac{1}{\sin\theta}-\sin\theta=\frac{\cos\theta}{\sin\theta}\cos\theta$$<br>
Taking the right-hand side of the equation.
$$=\frac{\cos^2\theta}{\sin\theta}$$<br>
$$=\frac{1-\sin^2\theta}{\sin\theta}$$<br>
$$=\frac{1}{\sin\theta}-\frac{\sin^2\theta}{\sin\theta}$$<br>
$$=\frac{1}{\sin\theta}-\sin\theta$$<br>
